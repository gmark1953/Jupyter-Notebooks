```python
# This document calculates the static H-Field at a point "P" along the Y-Axis 
# of a rectangular loop of length "L" (Z-Axis), width "W" (X-Axis), and "N", turns 
# referenced from the center of the loop.
# The field units are amps/meter.

from IPython.display import SVG
SVG(filename = 'CurrentLoop.svg')
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi

w = 0.5
l = 1
i = 3
n = 4

a = w/2
b = l/2
y = np.linspace(-1,1)

c = np.sqrt(a**2 + b**2 + y**2)
d = a**2 + y**2
e = b**2 + y**2
f = a**2 + b**2 + 2*y**2

hx = b*i*n*y/(pi*c*d)
hy = a*b*i*n*f/(pi*c*d*e)
hz = a*i*n*y/(pi*c*e)

plt.rcParams['figure.figsize'] = 8,5
plt.plot(y, hx, label='HX',linewidth=1.0)
plt.plot(y, hy, label='HY',linewidth=1.0)
plt.plot(y, hz, label='HZ',linewidth=1.0)
plt.xlabel('meters')
plt.ylabel('A/m')
plt.legend()
plt.title('H-Field VS Distance From Loop Center ')
plt.grid()
plt.axhline(y=0)
plt.axvline(x=0)
plt.show()
```
